```python
#Vytvořte program,ktery mi umožní vkládat další jména do tohoto listu. 
#Budu moct dovnitř vložit kolik jmen chci.
list_jmen =  ["Kamil", "Petr", "Vladimir", "Evzen","Valerie"]

while True:
  zadani = input(f"Vloz jmeno nebo dej (Exit) pro ukonceni. Aktualni pocet jmen: {len(list_jmen)}: ").title()

  if zadani == "Exit":
    print("Ukoncuji zadavani.")
    break

  elif zadani in list_jmen:
    print("Jmeno uz je v listu!")
    continue

  else
    print("Pridavam do listu.")
    list_jmen.append(zadani)
print(list_jmen)
```

```python
#Vytvorte pro každy program funkci
list_jmen =  ["Kamil", "Petr", "Vladimir", "Evzen","Valerie"]

def filtr_dle_jmena(vice=False):
  #Automatizuj program tak, abych si mohl zvolit na jake pismeno maji jmena zacinat.

  count = 0 #Promena count je dulezita pro vypsani: "nic jsem nenasel"
  pismeno = input("Zadej na jake pismeno mame zacinat!: ").title() #Zadavam a hlidam prvni pismeno

  if vice == False: #podminka resi o jaky mod se jedna
    for jmeno in list_jmen:
      if jmeno[0] == pismeno: #resim co zacina jmeno
        count += 1 #Pocitam kolik je tu jmen
        print(jmeno)
    if count == 0:
      print("Nic jsem nenasel")
  else: #vice je True
    for jmeno in list_jmen:
      if jmeno[:len(pismeno)] == pismeno:
        count += 1
        print(jmeno)
    if count == 0:
      print("Nic jsem nenasel")

  print(f"Nasel jsem celkem {count} jmen.")
```
