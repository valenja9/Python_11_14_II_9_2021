```python
#Z tohoto textu vytáhni jen jmena
text = "45667Jakub--------PEtr--------Kamil--------PEtr--------JoseF--------alfonc--------JARMILA--------Bára--------zita--------Xena(9333"
```

```python
text = text[5:-5]
```

```python
text
```

```python
upraveny_text = text.split("--------")
```

```python
#Vytvor program, ktery vsechna jmena prevede na stejny format... Karel, Jarmil, Xena
```

```python
#Jmena ulož do list_jmnen.
list_jmnen = []
for jmeno in upraveny_text:
  jmeno = jmeno.title()
  list_jmnen.append(jmeno)
print(list_jmnen)
```

```python
serazeny_list = [list_jmnen.pop()] 
for jmeno_s in list_jmnen:
  for index, jmeno in enumerate(serazeny_list):
    if jmeno_s < jmeno:
      serazeny_list.insert(index, jmeno_s)
      break
  else:
    serazeny_list.append(jmeno_s)
```

```python
import requests #Stahovani informaci znetu
import pandas as pd


r = requests.get("https://api.alternative.me/fng/?limit=1000")
data = r.json()
```

```python
data
```

```python
for d in data["data"]:
  print(d["value"], d["value_classification"]) #[""]
```

```python
from datetime import datetime

for d in data["data"]:
 d["timestamp"]=datetime.fromtimestamp(int(d["timestamp"]))
```

```python
df = pd.DataFrame(data["data"])
df
```

```python
import plotly.graph_objects as go
fig = go.Figure([go.Scatter(x=df['timestamp'], y=df['value'])])
fig.show()
```
